### Question
Given the following array $x$?

Write a single line of code that returns all of the odd numbers greater than 75?


### Starter Code

*HINT*: for computational efficiency the solution should use NumPy


````python
x = np.arange(100)
````

### Solution

<br>
<details>

  <summary>Click here to see the solution below</summary>
<br>

```


Any of the following lines are example solutions

y1 = np.intersect1d(np.where(x > 75),np.where(x%2!=0))
y2 = np.where((x > 75) & (x%2!=0))
y3 = np.where(np.logical_and(x > 75, x%2!=0))
y4 = x[(x > 75) & (x %2 !=0)]
```
